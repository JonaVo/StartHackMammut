```python
import cv2
import random

vidcap = cv2.VideoCapture("./data/vertical/Straight Wall (vertical).mp4")

startvec = []
endvec = []
startL = [15.054, 18.234, 21.274, 25.974, 30.454, 35.854, 40.354, 44.774, 49.374, 55.554, 61.514, 65.214, 68.494]
endL = [16.514, 19.714, 22.574, 27.674, 31.914, 37.414, 41.994, 46.334, 51.154, 57.914, 63.134, 66.254, 70.214]
startR = [13.754, 16.554, 23.694, 29.634, 33.574, 38.094, 42.914, 46.974, 51.974, 59.454, 62.594, 67.054, 71.614]
endR = [15.194, 17.894, 24.634, 31.214, 34.954, 39.714, 44.094, 48.614, 53.474, 61.074, 63.934, 68.214, 73.074]

startvec.extend(startR)
startvec.extend(startL)
startvec.sort()

endvec.extend(endR)
endvec.extend(endL)
endvec.sort()

randomvec = []
for i in range(len(endvec)):
    randomvec.append(random.randint(15, 72)+(random.randint(15,72)/100))
randomvec.sort()

for i in range(len(endvec)):
    name = "Photos/end/" + str(i)+".jpg"
    vidcap.set(cv2.CAP_PROP_POS_MSEC,endvec[i]*1000)      # just cue to 20 sec. position
    success,image = vidcap.read()
    if success:
        cv2.imwrite(name, image)     # save frame as JPEG file
        cv2.waitKey()  
        
for i in range(len(startvec)):
    name = "Photos/start/" + str(i)+".jpg"
    vidcap.set(cv2.CAP_PROP_POS_MSEC,startvec[i]*1000)      # just cue to 20 sec. position
    success,image = vidcap.read()
    if success:
        cv2.imwrite(name, image)     # save frame as JPEG file
        cv2.waitKey()  
        
for i in range(len(randomvec)):
    name = "Photos/random/" + str(i)+".jpg"
    vidcap.set(cv2.CAP_PROP_POS_MSEC,randomvec[i]*1000)      # just cue to 20 sec. position
    success,image = vidcap.read()
    if success:
        cv2.imwrite(name, image)     # save frame as JPEG file
        cv2.waitKey() 
        
for i in range(len(endL)):
    name = "Photos/leftarm/" + str(i)+".jpg"
    vidcap.set(cv2.CAP_PROP_POS_MSEC,endL[i]*1000)      # just cue to 20 sec. position
    success,image = vidcap.read()
    if success:
        cv2.imwrite(name, image)     # save frame as JPEG file
        cv2.waitKey() 
        
for i in range(len(endR)):
    name = "Photos/rightarm/" + str(i)+".jpg"
    vidcap.set(cv2.CAP_PROP_POS_MSEC,endR[i]*1000)      # just cue to 20 sec. position
    success,image = vidcap.read()
    if success:
        cv2.imwrite(name, image)     # save frame as JPEG file
        cv2.waitKey() 
```
